```python
from matplotlib import pyplot as plt
import pandas as pd
insurance = pd.read_csv('datas/insurance.csv')
```

```python
# 1
plt.scatter(insurance.bmi, insurance.charges, c=insurance.smoker, alpha=0.7, marker='^')

plt.xlabel("bmi")
plt.ylabel("charges")
cbar = plt.colorbar(ticks=[1,0])
cbar.set_ticklabels(['smoking','non-smoking'])

plt.show()
```

```python
# 2
num_bins = 12
plt.hist(insurance.charges[insurance.smoker==1],num_bins)
plt.hist(insurance.charges[insurance.smoker==0], num_bins, alpha=0.6)
plt.xlabel("Insurance Charges")
plt.ylabel("# of people")
plt.title("Distribution of Insurance Charge somking vs non-smoking")
plt.legend(['smoking', 'non-smoking'])
plt.show()
```

```python
# 3
num_bins = 15
plt.hist(insurance.charges[insurance.age<40],num_bins)
plt.hist(insurance.charges[insurance.age>=40], num_bins, alpha=0.7)
plt.xlabel("Insurance Charges")
plt.ylabel("# of people")
plt.title("Distribution of Insurance Charge by age <40 to >=40")
plt.legend(['< 40 years old', '>= 40 years old'])
plt.show()
```

```python
# 4
import numpy as np
num_bins = 15
# plt.hist(insurance.charges[insurance.age<40],num_bins)
# plt.hist(insurance.charges[insurance.age>=40], num_bins, alpha=0.7)
x = [0,1,2,3]
x_pos = [i for i, _ in enumerate(x)]
charge_sum_s = []
charge_sum_ns = []
for i in x:
  # charge_sum_s.append(insurance.loc[(insurance['smoker'] == 1) & (insurance['region'] == i), 'charges'].sum())
  # charge_sum_ns.append(insurance.loc[(insurance['smoker'] == 0) & (insurance['region'] == i), 'charges'].sum())

  charge_sum_s.append(np.sum(insurance.charges[insurance.smoker==1][insurance.region==i]))
  charge_sum_ns.append(np.sum(insurance.charges[insurance.smoker==0][insurance.region==i]))

# print(charge_sum_s)
# print(charge_sum_ns)
plt.bar(x_pos, charge_sum_s, alpha=0.8)
plt.bar(x_pos, charge_sum_ns, alpha=0.6)
plt.xlabel("Region")
plt.ylabel("Total paid")
plt.title("Region Insurance Charges by smoker/non-smoker")
plt.legend(['smoker', 'non-smoker'])
plt.xticks(np.arange(0,4,1))
plt.show()
```

```python
# 4.2
num_bins = 15
plt.hist(insurance.charges[insurance.sex==1],num_bins)
plt.hist(insurance.charges[insurance.sex==0], num_bins, alpha=0.7)
plt.xlabel("Insurance Charges")
plt.ylabel("# of people")
plt.title("Distribution of Insurance Charge by sex")
plt.legend(['Female', 'Male'])
plt.show()
```
