```python
from sklearn.datasets import load_iris
iris = load_iris()
attr = iris.data
print(iris.feature_names)
```

```python
# plt.scatter(attr[:,0], attr[:,1], s=80,c='g', marker='^')
plt.scatter(attr[:,0], attr[:,1], s=80,c='g', marker='^')

plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.show()
```

```python
target = iris.target

plt.scatter(attr[:,0], attr[:,1], s=100*attr[:,3]**2,c=target, alpha=0.5)

plt.xlabel("sepal length")
plt.ylabel("sepal width")
# plt.colorbar()
# plt.colorbar(ticks=[0,1,2])
# plt.colorbar(ticks=[0,1,1.7,2])
cbar = plt.colorbar(ticks=[0,1,2])
cbar.set_ticklabels(['Setosa','Versicolor','Virginica'])
plt.show()
```

```python
# histogram
target = iris.target
select_index = target == 0 # generate logical values on whether Setosa
setosa = attr[select_index,:]
num_bins = 25
plt.hist(setosa[:,1],num_bins)
plt.xlabel("sepal width")
plt.ylabel("number of observations")
plt.show()
```

```python
target = iris.target
select_index = target == 0 # generate logical values on whether Setosa
setosa = attr[target==0,:]
versicolor = attr[target==1,:]
num_bins = 15   #two histgram don't need to share the same bin size
plt.hist(setosa[:,1],num_bins,alpha=0.75)
plt.hist(versicolor[:,1],num_bins,alpha=0.75)
plt.xlabel("sepal width")
plt.ylabel("number of observations")
plt.legend(['Setosa', 'Versicolor'])
plt.show()
```
