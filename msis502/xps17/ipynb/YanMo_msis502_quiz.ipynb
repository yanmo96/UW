```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt

daily = pd.read_csv('datas/day.csv')
```

```python
# 1
plt.scatter(daily.temp, daily.cnt,c=daily.weathersit, alpha=0.6)

plt.xlabel("temp")
plt.ylabel("cnt")
cbar = plt.colorbar(ticks=[1,2,3])
cbar.set_ticklabels(['Clear','Mist','snow/rain'])
plt.show()
```

```python
# 2
plt.plot(daily.day, daily.casual, c='r')
plt.plot(daily.day, daily.cnt, 'b--')
plt.legend(['Casual', 'Total Checkout'])
plt.xlabel('Day')
plt.ylabel('Check Outs')
plt.show()
```
