```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
daily = pd.read_csv('datas/day.csv')
daily.head()
```

```python
# 1
# plt.plot(daily.iloc[:,0],daily.loc[:,'cnt'])
plt.plot(daily.day, daily.cnt)
plt.xlabel('Day')
plt.ylabel('Check Outs')
plt.show()
```

```python
# 2
plt.scatter(daily.temp, daily.casual, s=80,c='#539cab', marker='^', alpha=0.7)

plt.xlabel("temp")
plt.ylabel("casual cnt")
plt.show()
```

```python
# 3

plt.scatter(daily.temp, daily.casual, s=80,c=daily.workingday, marker='^', alpha=0.7)

plt.xlabel("temp")
plt.ylabel("casual cnt")
cbar = plt.colorbar(ticks=[0,1])
cbar.set_ticklabels(['non-working day','working day'])
plt.show()
```

```python
# 4

num_bins = 10
plt.hist(daily.registered,num_bins)
plt.xlabel("daily checkout")
plt.ylabel("count")
plt.title("Distribution of Registered Check Outs")
plt.show()
```

```python
# 5

num_bins = 5
plt.hist(daily.registered,num_bins, alpha=0.8)
plt.hist(daily.casual, num_bins, alpha=0.8)
plt.xlabel("daily checkout")
plt.ylabel("count")
plt.title("Distribution of Registered Check Outs")
plt.show()
```

```python
# 6
plt.scatter(daily.day, daily.temp,c=daily.season, alpha=0.5)

plt.xlabel("day")
plt.ylabel("temp")
cbar = plt.colorbar(ticks=[1,2,3,4])
cbar.set_ticklabels(['Winter','Spring','Summer','Fall'])
plt.show()
```

```python
# 6 (more options)
winter = daily.loc[daily.loc[:,'season'] == 1, 'temp']
spring = daily.loc[daily.loc[:,'season'] == 2, 'temp']
summer = daily.loc[daily.loc[:,'season'] == 3, 'temp']
fall = daily.loc[daily.loc[:,'season'] == 4, 'temp']

bin_size = 10
plt.hist(winter, bin_size, alpha=0.75)
plt.hist(spring, bin_size, alpha=0.75)
plt.hist(summer, bin_size, alpha=0.75)
plt.hist(fall, bin_size, alpha=0.75)
plt.legend(['Winter', 'Spring', 'Summer', 'Fall'])

plt.show()
```

```python
# 6 (more...)
import numpy as np
plt.hist(spring, bin_size, alpha=0.6)
plt.plot([np.mean(spring),np.mean(spring)],[0,40])
plt.plot([np.median(spring),np.median(spring)],[0,40],'--')
plt.show()
```

```python
# 7

plt.scatter(daily.day, daily.temp,c=daily.season, alpha=0.5, s=1000*daily.windspeed**3)

plt.xlabel("day")
plt.ylabel("temp")
cbar = plt.colorbar(ticks=[1,2,3,4])
cbar.set_ticklabels(['Winter','Spring','Summer','Fall'])
plt.show()
```
